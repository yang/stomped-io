```python
import re, pandas as pd
```

```python
suffix='2'
cpath = 'client'+suffix+'.log'
spath = 'server'+suffix+'.log'
```

```python
def extract(path, pat):
    return [
        map(float, m.groups())
        for line in open(path)
        for m in [re.match(pat, line)]
        if m
    ]
```

```python
cjank = pd.DataFrame(extract(cpath, r'client-jank: start (.+) end (.+) elapsed (.+)'),
                     columns=['start','end','elapsed'])
```

```python
cjank['gap'] = [0] + cjank.start.diff()
```

```python
cjank.describe()
```

```python
cjank['gap'].sort_values(ascending=False)[:10]
```
